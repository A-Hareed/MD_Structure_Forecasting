```python
def get_time_aromic_dataset(data_x,data_y,seq_length,t_func=timeseries_dataset_from_array):
    x = t_func(
        data=data_x,
        targets=None,
        sequence_stride = seq_length,
        sequence_length=seq_length,
        
    )

    y = t_func(
        data=data_y,
        targets=None,
        sequence_stride = seq_length,
        sequence_length=seq_length,
        
    )



    for input in x:
        print(input.shape)
        

    for input in y:
        print(input.shape)


    for batch in zip(x, y):
        inputs, targets = batch

    # targets = np.array(targets).reshape(targets.shape[0],(targets.shape[1]*targets.shape[2]))

    # print('target and features')
    # print(inputs.shape)
    # print(targets.shape)
    return inputs, targets
```
